## Reading Dataset:

```python
# Importing necessary libraries
import pandas as pd

# Load the dataset
df = pd.read_csv('Datasets.csv')  # Load the dataset into a Pandas DataFrame

# Display basic information about the dataset
df  # reading Dataset

```

## Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Drop unnecessary columns or columns causing issues
data_processed = df.drop(['ID', 'Sender_IP', 'Target_IP'], axis=1)  # Remove columns not needed for analysis

# Splitting features and target variable
X = data_processed.drop('class', axis=1)  # Features (independent variables)
y = data_processed['class']  # Target variable (dependent variable)

# Scale the features
scaler = StandardScaler()  # Initialize a scaler to standardize the features
X_scaled = scaler.fit_transform(X)  # Scale the features to have a mean of 0 and variance of 1

# Splitting data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
# Split the dataset into training (70%) and testing (30%) sets
```

##  Training, Prediction, and Evaluation :


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initialize the Random Forest Classifier model
model = RandomForestClassifier(n_estimators=100, random_state=42)  # Initialize Random Forest Classifier

# Train the model
model.fit(X_train, y_train) # Train the model using the training data

# Prediction on the test set
predict = model.predict(X_test)  # Make predictions on the test set

# Model Evaluation
accuracy = accuracy_score(y_test, predict)  # Calculate accuracy
print(f"Accuracy: {accuracy}")  # Print the accuracy of the model

# Additional evaluation metrics
print("Classification Report:")
print(classification_report(y_test, predictions))  # Display precision, recall, F1-score, and support for each class

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))  # Display the confusion matrix
```
